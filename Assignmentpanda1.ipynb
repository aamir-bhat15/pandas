Q5. How are DataFrame.size() and DataFrame.shape() different?

```python
DataFrame.size() and DataFrame.shape() are both methods in the Pandas library for working with DataFrames in Python. 
However, they have different purposes and return different values.

The DataFrame.size() method returns the total number of elements in the DataFrame. 
This includes all the rows and columns, so it is equal to the product of the number of rows and the number of columns in the DataFrame. 
The size includes both the actual data values as well as any NaN or missing values.

On the other hand, DataFrame.shape() returns a tuple that contains the dimensions of the DataFrame. 
It specifically returns the number of rows and the number of columns in the DataFrame. 
The shape attribute is helpful for understanding the structure and dimensions of a DataFrame without explicitly counting the elements.

In summary, DataFrame.size() gives the total number of elements in a DataFrame, 
while DataFrame.shape() returns a tuple with the number of rows and columns.
```

Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to
create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days
for each row in the DataFrame. The moving average should be calculated using a window of size 7 and
should include the current day.

```python
import pandas as pd

def calculate_moving_average(df, sales_column, date_column):
    df = df.sort_values(date_column)
    df['MovingAverage'] = df[sales_column].rolling(window=7, min_periods=1).mean()
    
    return df


df = pd.DataFrame({'Sales': [10, 15, 20, 12, 8, 25, 18, 16, 21],
                   'Date': ['2021-01-01', '2021-01-02', '2021-01-03', '2021-01-04', '2021-01-05', '2021-01-06', '2021-01-07', '2021-01-08', '2021-01-09']})

df['Date'] = pd.to_datetime(df['Date'])


df = calculate_moving_average(df, 'Sales', 'Date')


print(df)


'''

The function takes the DataFrame 'df', the name of the sales column, and the name of the date column as input. 
It first sorts the DataFrame by the date column in ascending order. 
Then it creates a new column 'MovingAverage' using the rolling() function with a window size of 7 and min_periods set to 1 to include the current day. 
Finally, it returns the updated DataFrame with the moving average calculated for each row.
'''
```

Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python
function to select all rows where the date is between '2023-01-01' and '2023-01-31'.

```python
import pandas as pd

df = pd.DataFrame({'Date': ['2023-01-01', '2023-01-02', '2022-01-03', '2021-01-04', '2021-01-05', '2021-01-06', '2023-01-07', '2023-01-08', '2023-01-09']})


```

```python
df[(df['Date']>'2023-01-01') & (df['Date']<='2023-01-31')]
```

```python

import pandas as pd

def select_rows_between_dates(df, date_column, start_date, end_date):
    # Convert start and end dates to datetime
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)
    
    # Select rows where date is between start and end dates
    selected_rows = df[(df[date_column] >= start_date) & (df[date_column] <= end_date)]
    
    return selected_rows

# Example usage:
df = pd.DataFrame({'Date': ['2023-01-01', '2023-01-15', '2023-01-25', '2023-02-05', '2023-02-10']})

# Convert the 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Call the function to select rows between '2023-01-01' and '2023-01-31'
selected_df = select_rows_between_dates(df, 'Date', '2023-01-01', '2023-01-31')

# Print the selected DataFrame
print(selected_df)
'''
The function takes the DataFrame 'df', the name of the date column, and the start and end dates as input. 
It first converts the start and end dates to datetime format using pd.to_datetime(). 
Then it uses boolean indexing with logical operators (&) to select rows where the date is greater than or equal to the start date and less than or equal to the end date. 
Finally, it returns a new DataFrame 'selected_rows' that contains only the selected rows.
'''
```
