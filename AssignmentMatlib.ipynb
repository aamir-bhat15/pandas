Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

```python
#a graph in which the values of two variables are plotted along two axes, the pattern of the resulting points revealing any correlation present.
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel('this is x axis')
plt.ylabel('this is y axis')
plt.title('this is scatter plot b/w x and y')
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

```python
#subplot() function adds subplot to a current figure at the specified grid position
import numpy as np
import matplotlib.pyplot as plt
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4= np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.title('First Graph')

plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.title('Second Graph')

plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.title('Third Graph')

plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.title('Forth Graph')

plt.tight_layout()
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar graph is a graphical representation of data in which we can highlight the category with particular shapes like a rectangle. It is used to represent the data in x and y axis

```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
#plt.figure(figsize(6,3))
plt.bar(company,profit)
plt.xlabel("Name of Companies")
plt.ylabel("Profit")
plt.title("Companies and corresponding profits")
plt.grid()
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

```python
Box plots are used to show distributions of numeric data values, especially when you want to compare them between multiple groups. 
They are built to provide high-level information at a glance, offering general information about a group of data's symmetry, skew, variance, and outliers.
```

```python
import numpy as np
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
fig,ax=plt.subplots()

ax.boxplot([box1,box2])

ax.set_xt
```
